```python
import pandas as pd
```

```python
df = pd.read_excel('dataset of PM dinner.xlsx')
```

```python
df.head(50)
```

```python
import numpy as np
df_filtered = df.dropna(subset=['category'])
df_result = df_filtered[['name', 'category', 'count', 'category2']]
df_result.head(10)
#df_result.to_csv('foodcategory.csv')
```

```python
category_counts = df_result.groupby('category')['count'].sum().reset_index()
```

```python
category_counts_sorted = category_counts.sort_values(by='count', ascending=False)
```

```python
category_counts_sorted.to_csv('foodcategory.csv')
```
